## Data Visualization

## Univariate Analysis: It means analysing single variable through Data visualization

```python
# here we are analysis sales column 
plt.figure(figsize=(10,6))
sns.distplot(data['Sales'],hist=False)
plt.title('Analysing Sales')
plt.xlabel('Sales')
plt.grid()
plt.show()
# here we can say that Sales is positive  or right skewness 
```

```python
# here we are analysing Profit column 
plt.figure(figsize=(12,8))
sns.distplot(data['Profit'],color='red')
plt.title('Analysing Profit')
plt.xlabel('Profit')
plt.grid()
plt.show()
# here we can say that Profit is positive also or right skewness 
```

```python
# here we are analysing State column which states that which state is saling maximum items
plt.figure(figsize=(12,8))
sns.countplot(y=data['State'])
plt.title('State has highest sales')
plt.ylabel('States')
plt.xlabel('Count of States appearing ')
plt.show()
#We can say that California is appering more as compare to all other states
```

```python
# we are analysing the sales through histogram 
plt.figure(figsize=(10,6))
bins=range(1000,23000,2000)
plt.hist(data['Sales'],bins=bins,color='purple')
plt.title('Sales Representation')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
# It  states that maximum sales is in 0-2500 rupes
```

```python
#represent Region of Country
data['Region'].value_counts().plot(kind='bar',color='orange',rot=0)
plt.title('Different regions')
plt.xlabel('Regions')
plt.ylabel('Frequency')
plt.show()
#here we can infer that west region has more sales
```

```python
#here we are analysing the percentage of category of products present in dataset 
plt.figure(figsize=(8,5))
plt.pie(data.Category.value_counts(),autopct='%.f%%',labels=['Office Supplies','Furniture','Technology'],shadow=True,explode=[0.1,0,0])
plt.title('Different categories')
plt.show()
#wecan say that office supplies are 60% present in data
```

```python
# we are finding labels for sub-categories
f=data['Sub-Category'].value_counts()
f.index
```

```python
#here we are analysing the percentage of subcategory of products of office-supplies present in dataset 
plt.figure(figsize=(10,5))
plt.pie(data['Sub-Category'].value_counts(),labels=f.index,autopct='%.f%%',shadow=True)
plt.title('Different categories')
plt.show()
#we can say that from office supplies blinders are mostly demanded
```

## Multi variate Analysing : It means analysing two variables or more variables together

```python
#Relationship between profit and discount
plt.figure(figsize=(10,6))
plt.scatter(data['Discount'],data['Profit'],color='purple')
plt.grid()
plt.show()
```

```python
#Relationship between profit and Sales
plt.figure(figsize=(12,8))
sns.scatterplot(data['Profit'],data['Sales'],color='pink',hue=data['Discount'],s=50)
plt.grid()
plt.show()
```
